```python
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

import os
import spotipy as sp
from spotipy.oauth2 import SpotifyClientCredentials

# Access environment variables
client_id = os.getenv("CLIENT_ID")
client_secret = os.getenv("CLIENT_SECRET")

client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")


sp = sp.Spotify(auth_manager = SpotifyClientCredentials(client_id = client_id, client_secret = client_secret))


favorite_artist_name = "Beyonce"
beyonce_id = "6vWDO969PvNqNYHIOW5v0m"

response = sp.artist_top_tracks("6vWDO969PvNqNYHIOW5v0m")

tracks = response['tracks']
top_tracks = sp.artist_top_tracks(beyonce_id)


print(f"Top 10 tracks of {favorite_artist_name}:")
for track in top_tracks['tracks'][:10]:
        print(f"{track['name']} - {', '.join([artist['name'] for artist in track['artists']])}")
else:
    print(f"Artist '{favorite_artist_name}' not found.")
```

```python
import pandas as pd

tracks_df = pd.DataFrame(tracks)
sorted_tracks = sorted(tracks, key=lambda x: x['popularity'], reverse=True)

# Print top 3 tracks by popularity
print("Top 3 Tracks by Popularity:")
for i, track in enumerate(sorted_tracks[:3], start=1):
    print(f"{i}. {track['name']} - Popularity: {track['popularity']}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
plt.scatter(tracks_df['duration_ms'], tracks_df['popularity'], color='blue', alpha=0.7)
plt.title('Song Duration vs. Popularity')
plt.xlabel('Duration (seconds)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

There is no correlation between duration and popularity.
